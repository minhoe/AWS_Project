### Model Definition

```python
# Network Topologies
n_hidden_1 = 521
n_hidden_2 = 384
n_hidden_3 = 256

n_input    = 784
n_classes  = 10
display_step = 5

class VAE_Model:
    def __init__(self, sess, name, learning_rate=1e-3, batch_size=100, n_z=10):
        self.sess = sess
        self.name = name
        
        self.learning_rate = learning_rate
        self.batch_size    = batch_size
        self.n_z           = n_z
        
        self.Build()
        self.sess.run(tf.global_variables_initializer())
    
    # Build model and loss function
    def Build(self):
        # Layers
        with tf.variable_scope(self.name):
            # Set Inputs and Outputs
            self.x = tf.placeholder(dtype=tf.float32, shape=[None, n_input])
            
            # Set Layers - Encode
            # x -> (z_mean, z_sigma) -> z
            g_zx_1 = tf.layers.Dense(units=n_hidden_1, activation=tf.nn.elu)(self.x)
            g_zx_2 = tf.layers.Dense(units=n_hidden_2, activation=tf.nn.elu)(g_zx_1)
            g_zx_3 = tf.layers.Dense(units=n_hidden_3, activation=tf.nn.elu)(g_zx_2)
            
            self.z_mu           = tf.layers.Dense(units=self.n_z, activation=None)(g_zx_3)
            self.z_log_sigma_sq = tf.layers.Dense(units=self.n_z, activation=None)(g_zx_3)

            eps = tf.random_normal(shape=tf.shape(self.z_log_sigma_sq),
                                   mean=0, stddev=1, dtype=tf.float32)
            self.z = self.z_mu + tf.sqrt(tf.exp(self.z_log_sigma_sq)) * eps
           
            # Set Layers - Decode
            # z -> x_hat
            g_xz_1 = tf.layers.Dense(units=n_hidden_3, activation=tf.nn.elu)(self.z)
            g_xz_2 = tf.layers.Dense(units=n_hidden_2, activation=tf.nn.elu)(g_xz_1)
            g_xz_3 = tf.layers.Dense(units=n_hidden_2, activation=tf.nn.elu)(g_xz_2)
            self.x_hat = tf.layers.Dense(units=n_input, activation=tf.sigmoid)(g_xz_3)
            
        # Error : 
        # 1) Reconstruction Loss (Binary Cross Entropy)
        # BCE(x, x_hat) = (1/N) * SUM[ x*log(x_hat) + (1-x)*log(1-xhat)]
        ep = 1e-10
        rec_loss = -tf.reduce_sum(
            self.x * tf.log(ep + self.x_hat) + (1-self.x) * tf.log(ep + (1-self.x_hat)),
            axis=1
        )
        self.rec_loss = tf.reduce_mean(rec_loss)
        
        # 2) Regularization loss
        # KLD( P(z_mu, z_sigma_sq) || N(0, 1))
        reg_loss = -0.5 * tf.reduce_sum(
            1 + self.z_log_sigma_sq - tf.square(self.z_mu) - tf.exp(self.z_log_sigma_sq),
            axis=1
        )
        self.reg_loss = tf.reduce_mean(reg_loss)
        
        self.tot_loss = tf.reduce_mean(rec_loss + reg_loss)
        self.optm = tf.train.AdamOptimizer(
            learning_rate=self.learning_rate).minimize(self.tot_loss)
        print('Model definition completed') # Check the grammar error above
    
    # Execute the forward and backward process with batch dataset
    def run_single_step(self, batch_xs):
        _, loss, rec_loss, reg_loss = sess.run(
                        [self.optm, self.tot_loss, self.rec_loss, self.reg_loss],
                        feed_dict={self.x:batch_xs})
        return loss, rec_loss, reg_loss
    
    # x -> x_hat
    def reconstructor(self, x):
        x_hat = self.sess.run(self.x_hat, feed_dict={self.x: x})
        return x_hat
    
    # z -> x_hat
    def generator(self, z):
        # Note that you can replace any tensor in your computation graph using feed_dict 
        # -- it's not restricted to just placeholders!
        x_hat = self.sess.run(self.x_hat, feed_dict={self.z: z}) 
        return x_hat
    
    # x -> z
    def transformer(self, x):
        z = self.sess.run(self.z, feed_dict={self.x: x})
        return z
```

```python
def trainer(train_images,
            sess, name, learning_rate=1e-3, batch_size=100, training_epochs=100, n_z=10):
    
    N_TRN = len(train_labels)
    
    # Data Preprocessing
    train_images      = train_images / 255.0
    train_images_flat = sess.run(tf.reshape(train_images, shape=(N_TRN,-1)))
    print(train_images_flat.shape)
    
    model = VAE_Model(sess, name, learning_rate=learning_rate, batch_size=batch_size, n_z=n_z)

    # Minibatch learning    
    for epoch in range(training_epochs): 
        num_batch = int(N_TRN / batch_size)
        
        avg_tot_loss = 0.
        avg_rec_loss = 0.
        avg_reg_loss = 0.
        
        # Get random minibatch for each epoch
        randindices = np.random.permutation(train_images_flat.shape[0])

        for i in range(num_batch):
            # Obtain a batch
            cur_indices = randindices[i*batch_size:(i+1)*batch_size]
            batch_xs    = train_images_flat[cur_indices, :]
            
            # Execute the forward and backward process and compute losses
            tot_loss, rec_loss, reg_loss = model.run_single_step(batch_xs)
            avg_tot_loss += tot_loss / num_batch
            avg_rec_loss += rec_loss / num_batch
            avg_reg_loss += reg_loss / num_batch

            # Display training steps
        if epoch % display_step == 0:
            print('Epoch: %03d/%03d cost: %0.9f rec_cost: %.3f reg_cost %.3f'
                  % (epoch, training_epochs, avg_tot_loss, avg_rec_loss, avg_reg_loss))

    print('Done')
    return model
```

### Main function

```python
%%time
NAME = 'FashionMNIST'

# Session
tf.reset_default_graph()
sess = tf.Session(config=tf.ConfigProto(log_device_placement=True))

# Train the model
model = trainer(train_images,
                sess, NAME,
                learning_rate=1e-4,  batch_size=100, training_epochs=100, n_z=10)
```

```python
# Network Topologies
n_hidden_1 = 128
n_input    = 784
n_classes  = 10
    
# Inputs and Outputs
x = tf.placeholder(tf.float32, shape=[None, n_input], name='X')
y = tf.placeholder(tf.float32, shape=[None, n_classes], name='Y')

#NETWORK PARAMETERS
stddev = 0.1
weights = {
    'h1' : tf.Variable(tf.random_normal(shape=[n_input,    n_hidden_1], stddev=stddev)),
    'out': tf.Variable(tf.random_normal(shape=[n_hidden_1, n_classes],  stddev=stddev))
}

bias = {
    'b1' : tf.Variable(tf.random_normal(shape=[n_hidden_1])),
    'out': tf.Variable(tf.random_normal(shape=[n_classes]))
}

print('Model definition completed') # Check the grammar error above
```

### MLP as a function

```python
def MLP(X_, weights_, bias_):
    layer_1 = tf.nn.relu(tf.add(tf.matmul(X_, weights_['h1']), bias_['b1']))
    out     = tf.add(tf.matmul(layer_1, weights_['out']), bias_['out'])
    return out
```

### Create Tensor graph for ANN

```python
# Prediction
pred = MLP(x, weights, bias)

# Loss and Optimizer
# Do calculate C/E with logits.. not with softmax values
cost = tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred, labels=y)
cost = tf.reduce_mean(cost)

#cost = tf.reduce_mean(-tf.reduce_sum(y*tf.log(tf.nn.softmax(pred) + 0.1), reduction_indices=1))

optm = tf.train.AdamOptimizer(learning_rate=0.0001).minimize(cost)

corr = tf.equal(tf.argmax(pred, 1), tf.argmax(y, 1))
accr = tf.reduce_mean(tf.cast(corr, tf.float32))

# Initializer
init = tf.global_variables_initializer()

print('Tensor graph completed')

# Session
sess = tf.Session(config=tf.ConfigProto(log_device_placement=True))
sess.run(tf.global_variables_initializer())
```

```python
N_CLS = 10
N_TRN = len(train_labels)
N_TST = len(test_labels)

train_labels_1h = sess.run(tf.one_hot(train_labels, depth=10))
test_labels_1h  = sess.run(tf.one_hot(test_labels, depth=10))

train_images_flat = sess.run(tf.reshape(train_images, shape=(N_TRN,-1)))
test_images_flat  = sess.run(tf.reshape(test_images, shape=(N_TST,-1)))

print(train_images_flat.shape)
print(test_labels_1h.shape)
```

### Fit the model

```python
training_epochs = 20
batch_size      = 32
display_step    = 5

# Minibatch learning
for epoch in range(training_epochs): # Epochs
    avg_cost = 0.0
    num_batch = int(N_TRN / batch_size)
    
    # Get random minibatch for each epoch
    randindices = np.random.permutation(len(train_labels_1h))
    
    for i in range(num_batch):
        cur_indices = randindices[i*batch_size:(i+1)*batch_size]
    
        batch_xs = train_images_flat[cur_indices, :]
        batch_ys = train_labels_1h[cur_indices, :]
    
        feeds = {x:batch_xs, y:batch_ys}
    
        sess.run(optm, feed_dict=feeds)
        avg_cost += sess.run(cost, feed_dict=feeds) / num_batch
    
    # Display training steps
    if epoch % display_step == 0:
        feeds_train = {x: batch_xs, y: batch_ys}
        feeds_test  = {x: test_images_flat, y:test_labels_1h}
        
        train_acc = sess.run(accr, feed_dict=feeds_train)
        test_acc  = sess.run(accr, feed_dict=feeds_test)
        
        print('Epoch: %03d/%03d cost: %0.9f train_acc: %.3f test_acc %.3f'
              % (epoch, training_epochs, avg_cost, train_acc, test_acc))

print('Done')
```
